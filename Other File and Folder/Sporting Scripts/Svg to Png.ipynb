```python


"""
Instrunction to run this code
download this file and install 

 https://sk1project.net/uc2/download/ 

go to C:\Program Files\UniConvertor-2.0rc5\dlls


and add this to  windows env path 


"""

import os
import cairosvg

# Define the input and output directories
input_dir = r'C:\Users\admin\Django\Batch 420\data\logo\simple-svg-brand-logos\logos'
output_dir = r'C:\Users\admin\Django\Batch 420\data\logo\simple-svg-brand-logos\png_logos'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Loop through each file in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".svg"):
        file_path = os.path.join(input_dir, filename)
        
        # Create the new filename with a .png extension
        new_filename = os.path.splitext(filename)[0] + ".png"
        output_path = os.path.join(output_dir, new_filename)
        
        # Convert the SVG to PNG and save it
        cairosvg.svg2png(url=file_path, write_to=output_path)
        
        print(f"Converted {filename} to {new_filename} and saved to {output_path}")

```
