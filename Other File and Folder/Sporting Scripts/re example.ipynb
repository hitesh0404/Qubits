```python

import re
def rep(old):
    old=str(old)
    new='href="{% static \''+old[6:-1]+'\' %}"'
    return  str(new)

def href_replace(address,pattern=r'''href="([^"]*)"'''):
    with open(address,'r+') as file:
        text = file.read()
        new_text=''
        count=0
        extracted_re_iterable=re.finditer(pattern,text)
        previous_end=0
        for i in extracted_re_iterable:
            count+=1
            temp_to=rep(i.group())
            new_text=new_text+text[previous_end:i.start()]+temp_to
            previous_end=i.end()
        new_text=new_text+text[previous_end:]
        file.seek(0)
        file.write(new_text)
    return new_text
```
